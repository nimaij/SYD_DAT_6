```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Location = r'/Users/jiaminlim93/Desktop/SYD_DAT_6/labs/Week 1/u.user'
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_table(Location, sep='|', header=None, names=user_cols, index_col='user_id', dtype={'zip_code':str})
```

```python
users.head()
```

```python
# for each occupation in 'users', count the number of occurrences
users.occupation.value_counts()
```

```python
type(users)
```

```python
# for each occupation, calculate the mean age
age_mean_by_occupation = users.groupby('occupation')['age'].mean()
age_mean_by_occupation
```

```python
# for each occupation, calculate the minimum ages
age_min_by_occupation = users.groupby('occupation')['age'].min()
age_min_by_occupation
```

```python
# for each occupation, calculate the maximum ages
age_max_by_occupation = users.groupby('occupation')['age'].max()
age_max_by_occupation
```

```python
# for each combination of occupation and gender, calculate the mean age
age_mean_by_gender_and_occupation = users.pivot_table('age','occupation','gender','mean')
```

```python
age_mean_by_gender_and_occupation
```

```python
# randomly sample a DataFrame
sampler = np.random.permutation(len(users))
users.take(sampler[:5]) # taking a random sample of 5
```

```python
# detect duplicate users
users.duplicated()
users.drop_duplicates()
# The dataframe was reduced from 943 users to 936 users. There were 7 duplicate entries.
```
